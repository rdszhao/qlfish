```python
def generate_query(parsed):
	try:
		if 'subq' in parsed:
			subq = parsed['subq']
			if type(subq) == str:
				parsed['tables'] = [subq]
				del parsed['subq']
			else:
				parsed['subq'] = generate_query(parsed['subq'])
		return db.query(**parsed)
	except:
		return db.query(**parsed)

def qquery(**kwargs):
    return kwargs

def generate_qquery(parsed):
	try:
		if 'subq' in parsed:
			subq = parsed['subq']
			if type(subq) == str:
				parsed['tables'] = [subq]
				del parsed['subq']
			else:
				parsed['subq'] = generate_qquery(parsed['subq'])
		return qquery(**parsed)
	except:
		return qquery(**parsed)
```
